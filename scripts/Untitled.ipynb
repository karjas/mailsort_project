```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
df = pd.read_csv("../data/train.csv")
lab = df["label"].values
```

```python
print(df['org'])
```

```python
print(df.columns)
org = df['org'].fillna("nan").values.reshape(-1,1)
print(org)
```

```python
enc = OneHotEncoder(handle_unknown = 'ignore')
enc.fit(org)
```

```python
W = enc.transform(org)
```

```python
X_tr, X_te, Y_tr, Y_te = train_test_split(W,lab, test_size = 0.2)
```

```python
clf = svm.SVC(kernel='rbf', C = 5).fit(X_tr, Y_tr)
```

```python
clf.score(X_te, Y_te)
```

```python
d_test = pd.read_csv("../data/test.csv")
```

```python
yy = d_test['org'].fillna("nan").values.reshape(-1,1)
```

```python
pred_y = enc.transform(yy)
```

```python
pr = clf.predict(pred_y)
pr = [int(p) for p in pr]
print(pr)
```

```python
pred_dv = pd.DataFrame(pr, columns=['label'])
pred_dv.to_csv("org_submission.csv", index = True, index_label = 'Id')
```
